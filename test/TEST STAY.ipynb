```python
%%writefile test_units.py

# importing the sibling folder is not as easy as it should be..
import os, sys
here = os.path.split(os.path.abspath(os.path.dirname(__file__)))
src = os.path.join(here[0], "src")
sys.path.append(src)

from stay import loads


def test_docs():
    text = ""
    assert list(loads(text.splitlines())) == [{}]
    
    text = """
===
===
"""
    assert list(loads(text.splitlines())) == [{}, {}, {}]
    
def test_dict():
    text = """
a: b
b: a
"""
    assert list(loads(text.splitlines())) == [{'a': 'b', 'b': 'a'}]

    
def test_comments():
    text = "# adsf"
    assert list(loads(text.splitlines())) == [{}]
    
    text = """
###
asdf
###
a: b
"""
    assert list(loads(text.splitlines())) == [{'a': 'b'}]
    
    text = """
### asdf ###
a: b
"""
    assert list(loads(text.splitlines())) == [{'a': 'b'}]
    
def test_long():
    text = """
foo:::
1
2
3


:::   
"""
    assert list(loads(text.splitlines())) == [{'foo': '1\n2\n3\n\n'}]

def test_complex():
    text = """
name: 
    family: adsf
    call: 
        foo:::
1
2
:::    
"""
    assert list(loads(text.splitlines())) == [{'name': {"family": "adsf", "call": {"foo": "1\n2"}}}]
    
def test_list():
    text = """
matrix:::[

[1 2 3]
[4 5 6]
[7 8 9]
foo bar
]:::
"""
    assert list(loads(text.splitlines())) == [{"matrix": [['1','2','3'], ['4','5','6'], ['7','8','9'], "foo bar"]}]
```

```python
%%writefile test_functionality.py

import os, sys
here = os.path.split(os.path.abspath(os.path.dirname(__file__)))
src = os.path.join(here[0], "src")
sys.path.append(src)

from stay import dumps, loads, load
from pydantic.dataclasses import dataclass
from typing import List
from dataclasses import asdict

d = {"2": """asdf\nadsf"""}
assert d == list(loads(dumps(d).splitlines()))[0]

@dataclass
class Foo:
    num: int
    foo: str
    bar: str
    baz: List[int]

D1 = {"num":1, "foo": "qwer", "bar": "asdf\nadsf", "baz": [1,2,3]}

with open("test", "w") as f:
    f.write(dumps(D1))
    
with open("test") as f:
    for x in load(f):
        D2 = asdict(Foo(**x))

assert D1 == D2
```
